## Cartopy

A simple all-in-one visualization package that has built-in vector and raster information and projection systems (`crs` - Coordinate Reference System).

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

# For this demo, we choose the Plate Carre projection which is straightforward and preserves latitude and longitude lines as straight lines.
# It does suffer from distortions in terms of area and shape as you move away from the equator. (The scale is true only along the equator).
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Note that `ax` is an object from cartopy, not matplotlib,
# so it has lots of fancy GIS related methods available

# Rough extents of Africa
#   20W to 60E longitude, 40S to 45N latitude
# Remember the order and signs
#   Longitudes, then Latitudes
#   - to + as to proceed east of Greenwich
#   - to + as you proceed north of equator
ax.set_extent([-20, 60, -40, 45], crs=ccrs.PlateCarree())

# "Features" have a sensible default color - yellow for land, blue for oceans
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle='-')

ax.add_feature(cfeature.RIVERS)

# We can override feature properties, of course
ax.add_feature(cfeature.LAKES, alpha=0.9, color='orange')
```
