```python
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader
import matplotlib.pyplot as plt
```

```python
# Vector files for Angola (shapefiles) downloaded from gadm.org
# 4 shapefiles are provided, (0-3), which progressively incorporate
# more and more administrative boundaries within the country.
# We'll just deal with the 0th to keep things simple
filename = '../src/deepmap/data/gadm41_AGO_shp/gadm41_AGO_0.shp'

# Cartopy has many readers for different file extensions
reader = shpreader.Reader(filename)
# `shapes` is a list of MultiPolygon objects
shapes = list(reader.geometries())

ax = plt.axes(projection=ccrs.PlateCarree())
# Note that `ax` is an object from cartopy, not matplotlib,
# so it has lots of fancy GIS related methods available
ax.coastlines(resolution='10m')
# Add custom shapes to our map; specify projections for which
# these shapes are defined
ax.add_geometries(geoms=shapes, crs=ccrs.PlateCarree(),
                  edgecolor='black', facecolor='pink', alpha=1)

ax.set_extent([0, 30, -40, 40], crs=ccrs.PlateCarree())

plt.title('Angola')
```
